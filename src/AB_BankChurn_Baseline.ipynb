## Putting together all the Preprocessing steps that needs to be followed before using the data
- Numerical Scaling of CreditScore, Age, Tenure, Balance, Estimated Salary
- Categorical numbering/mapping of Geography, Gender
- This preprocessing includes both the train and test data

```python
def dataPreprocessing(df):
    # scalar properties
    df['CreditScore'] = MinMaxScaler().fit_transform(df[['CreditScore']])
    df['Age'] = MinMaxScaler().fit_transform(df[['Age']])
    df['Balance'] = MinMaxScaler().fit_transform(df[['Balance']])
    df['EstimatedSalary'] = MinMaxScaler().fit_transform(df[['EstimatedSalary']])
    df['Tenure'] = MinMaxScaler().fit_transform(df[['Tenure']])

    # categorical propeties
    mapping_dict = {'France': 0, 'Spain': 1, 'Germany': 2}
    df['Geography'] = df['Geography'].map(mapping_dict)

    mapping_dict = {'Male': 0, 'Female': 1}
    df['Gender'] = df['Gender'].map(mapping_dict)

    return df
    
    
```

## Logistic Regression with Scikit Learn

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
modified_df = dataPreprocessing(train_csv)
trainDf, validDf = train_test_split(modified_df, test_size=0.33, random_state=42)
print("Train Df length : ", len(trainDf))
print("Test Df length : ", len(validDf))
```

```python
trainDf.head()
```

```python
training_columns  = ['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
```

```python
X_train = trainDf[training_columns].values
```

```python
print("X train : \n")
print(X_train)
```

```python
print("X train shape : ", X_train.shape)
```

```python
Y_train = trainDf[['Exited']].values
print("Y train shape : ", Y_train.shape)
```

```python
Y_train = Y_train.reshape((len(Y_train)))
print(Y_train.shape)
```

```python
# Defining the model
model = LogisticRegression(solver='liblinear', max_iter=500, random_state=0)
model.fit(X_train, Y_train)
```

```python
print("Model Classes")
print(model.classes_)
```

```python
model.predict_proba(X_train)
```

```python
model.predict(X_train)
```

```python
print("Training accuracy of the model")
model.score(X_train, Y_train)
```

```python
X_valid = validDf[training_columns].values
Y_valid = (validDf[['Exited']].values).reshape((len(validDf)))

print("X valid shape : ", X_valid.shape)
print("Y valid shape : ", Y_valid.shape)
```

```python
model.score(X_valid, Y_valid)
```

```python
final_prob_predictions = []
for i in range(len(prob_predictions)):
    if(prob_predictions[i][0] > prob_predictions[i][1]):
        final_prob_predictions.append(prob_predictions[i][0])
    else:
        final_prob_predictions.append(prob_predictions[i][1])
        
print(final_prob_predictions)
```
